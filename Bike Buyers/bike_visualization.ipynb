```python
import pandas as pnds
import numpy as nmpy
import seaborn as sbrn
import matplotlib.pyplot as mplt

# Importing cleaned data set
ds =  pnds.read_csv("bike_buyers_clean.csv")
```

```python
# Change purchased bike value into int
ds['Purchased Bike'] = ds['Purchased Bike'].map({'Yes' : 1.0, 'No' : 0.0})   
```

```python
#create chart according to education
fig, bx=mplt.subplots(figsize=(10, 5))
visualA=sbrn.barplot( data = ds,x = 'Education', y = 'Purchased Bike', ci=0,palette="viridis",saturation=10)

#physical property adjustment of the graph
visualA.set(title="Education Based Bike Purchase Graph")
visualA.xaxis.get_label().set_fontsize(18)
visualA.yaxis.get_label().set_fontsize(18)
visualA.title.set_fontsize(24)
```

```python
#create chart according to income
fig, ax=mplt.subplots(figsize=(20, 5))
sbrn.set_theme(style='darkgrid', palette=None)

#color and visual adjutments
visualB=sbrn.histplot(data=ds,x="Income", hue="Purchased Bike" ,shrink=0.8,kde=True,
             line_kws= {'linestyle': 'dashed'},multiple="stack", element="poly",
             palette=sbrn.blend_palette(['purple','blue'] ,2))

#physical property adjustment of the graph
visualB.set(title="Income Based Bike Purchase Graph")
visualB.xaxis.get_label().set_fontsize(18)
visualB.yaxis.get_label().set_fontsize(18)
visualB.title.set_fontsize(24)

#display
mplt.show()
```

```python
#create chart according to occuption
fig, cx=mplt.subplots(figsize=(10, 5))
visualC=sbrn.barplot(x = 'Purchased Bike', y = 'Occupation', data = ds, palette='twilight_shifted_r',ci=0)

#physical property adjustment of the graph
visualC.set(title="Occuption Based Bike Purchase Graph")
visualC.xaxis.get_label().set_fontsize(18)
visualC.yaxis.get_label().set_fontsize(18)
visualC.title.set_fontsize(24)
```

```python
#create chart according to age
displot = ds['Age']
visualD=sbrn.displot(data=ds, x=displot, kind='kde', rug=True,fill=True, height=5, color="darkmagenta")

#title adjustment
visualD.set(title="Age Based Bike Purchase Graph")
```

```python
#create chart according to region
fig, ax=mplt.subplots(figsize=(10, 5))
visualE=sbrn.countplot(x="Purchased Bike" ,data=ds, hue='Region', palette="icefire", saturation=0.7)

#physical property adjustment of the graph
visualE.set(title="Region Based Purchase Graph",ylabel='Count')
visualE.xaxis.get_label().set_fontsize(18)
visualE.yaxis.get_label().set_fontsize(18)
visualE.title.set_fontsize(24)

#display
mplt.show()
```

```python
#create and resize chart that shows age vs gender
fig, dx=mplt.subplots(figsize=(10, 5))
visualF=sbrn.histplot(x = 'Age', hue='Gender', data = ds,shrink=0.8,multiple="dodge",palette="ocean")
```
